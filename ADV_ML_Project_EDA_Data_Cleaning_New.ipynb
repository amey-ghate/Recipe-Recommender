```python
#%pip install pandas
import pandas as pd
import numpy as np

data_reviews = pd.read_csv('Reviews Data.csv')

data_reviews.head(10)
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt
from nltk.util import ngrams
import string

# Sample code to load the stopwords and punkt data (if not already downloaded)
nltk.download('stopwords')
nltk.download('punkt')

# Assuming you have a DataFrame named data_reviews with a "Review" column
# Tokenize, lowercase, and remove stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

# Function to clean text by removing stopwords and punctuation
def clean_text(text):
    words = word_tokenize(text.lower())
    return [word for word in words if word not in stop_words and word not in punctuation]

# Apply the cleaning function to each review
data_reviews['Review'] = data_reviews['Review'].apply(lambda x: ' '.join(clean_text(x)))

# Combine all reviews into a single string
all_reviews = ' '.join(data_reviews['Review'])

# Tokenize the combined text into words
words = word_tokenize(all_reviews)

# Create bigrams (pairs of adjacent words)
bigrams = list(ngrams(words, 2))

# Count bigram frequencies
bigram_freq = Counter(bigrams)

# Get the top 20 common bigrams and their frequencies
top_bigrams = bigram_freq.most_common(20)

# Format bigrams for plotting (convert tuples to strings)
top_bigrams_formatted = [(f'{bigram[0]} {bigram[1]}', freq) for bigram, freq in top_bigrams]

# Plot the top 20 common bigrams
top_bigrams_df = pd.DataFrame(top_bigrams_formatted, columns=['Bigram', 'Frequency'])
plt.figure(figsize=(12, 6))
plt.barh(top_bigrams_df['Bigram'], top_bigrams_df['Frequency'])
plt.xlabel('Frequency')
plt.ylabel('Bigram')
plt.title('Top 20 Common Bigrams in Reviews (Excluding Punctuation)')
plt.gca().invert_yaxis()  # Invert the y-axis to show the most common bigram at the top
plt.show()

```

DATA RECIPES EDA


```python
import pandas as pd
import numpy as np
data_recipes = pd.read_csv('Recipes Data (1).csv')

data_recipes.columns
```

```python
data_recipes
```

```python
data_recipes = data_recipes.dropna()
```

```python
data_recipes.isnull().sum()
```

```python
data_recipes['DatePublished'] = pd.to_datetime(data_recipes['DatePublished'])
data_recipes['DatePublished'] = data_recipes['DatePublished'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
data_recipes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Get the top 20 recipe categories by count
top_recipe_categories = data_recipes['RecipeCategory'].value_counts().head(20)

# Count plot to visualize the number of recipes in the top 20 categories
plt.figure(figsize=(12, 6))
sns.barplot(x=top_recipe_categories.values, y=top_recipe_categories.index, orient='h')
plt.xlabel('Number of Recipes')
plt.ylabel('Recipe Category')
plt.title('Top 20 Recipe Categories by Count')
plt.show()
```

```python
# Convert 'DatePublished' to datetime
data_recipes['DatePublished'] = pd.to_datetime(data_recipes['DatePublished'])

# Create a time series plot for recipe publications over time
plt.figure(figsize=(12, 6))
data_recipes['DatePublished'].value_counts().sort_index().plot()
plt.xlabel('Date Published')
plt.ylabel('Number of Recipes Published')
plt.title('Recipe Publications Over Time')
plt.grid(True)
plt.show()
```

```python
from wordcloud import WordCloud

# Generate a word cloud from the 'Keywords' column
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(data_recipes['Keywords']))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Recipe Keywords')
plt.show()
```
